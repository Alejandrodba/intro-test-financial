ta: es una biblioteca que proporciona una amplia variedad de indicadores técnicos para el análisis de estrategias de trading. Algunos de los indicadores que se pueden calcular con ta incluyen el RSI, el MACD, el promedio móvil, entre otros.

backtrader: es una biblioteca de backtesting de estrategias de trading. Permite simular estrategias de trading en datos históricos y evaluar su desempeño.

zipline: es una biblioteca de backtesting y trading algorítmico. Permite la creación de estrategias de trading y su ejecución en tiempo real o en datos históricos.

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
from ta import add_all_ta_features
from ta.utils import dropna
from ta.volatility import BollingerBands
```

```python
df = yf.download(tickers = "SPY",  # list of tickers
            period = "50y",         # time period
            interval = "1d",       # trading interval
            ignore_tz = True,      # ignore timezone when aligning data from different exchanges?
            prepost = False)       # download pre/post market hours data?
```

```python
# Clean NaN values
df = dropna(df)

# Initialize Bollinger Bands Indicator
indicator_bb = BollingerBands(close=df["Close"], window=20, window_dev=2)

# Add Bollinger Bands features
df['bb_bbm'] = indicator_bb.bollinger_mavg()
df['bb_bbh'] = indicator_bb.bollinger_hband()
df['bb_bbl'] = indicator_bb.bollinger_lband()

# Add Bollinger Band high indicator
df['bb_bbhi'] = indicator_bb.bollinger_hband_indicator()

# Add Bollinger Band low indicator
df['bb_bbli'] = indicator_bb.bollinger_lband_indicator()
```
